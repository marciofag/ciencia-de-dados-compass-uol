# Tópicos do Desafio 1

## Quais são as categorias com maior e menor receita dos últimos 12 meses?

```python
# Merge entre itens_pedido, produtos, pedidos e pagamentos_pedido
merge_1 = pd.merge(itens_pedido, produtos, on='product_id', how='left')
merge_categoria = pd.merge(merge_1, pedidos, on='order_id', how='left')

# Vamos supor que a coluna 'order_purchase_timestamp' contenha a data da compra
# E que estamos considerando os últimos 12 meses a partir da data mais recente no conjunto de dados
data_recente = pd.to_datetime(merge_categoria['order_purchase_timestamp']).max()
ano_passado = data_recente - pd.DateOffset(months=12)

# Filtrar as transações nos últimos 12 meses
transacoes_recentes = merge_categoria[pd.to_datetime(merge_categoria['order_purchase_timestamp']) >= ano_passado]

transacoes_recentes.head()
```

```python
# Calcular a receita total por categoria nos últimos 12 meses
receita_por_categoria = transacoes_recentes.groupby('product_category_name')['price'].sum().reset_index()
top_receita_por_categoria = receita_por_categoria.sort_values(by='price', ascending=False)
print(top_receita_por_categoria)
```

```python
# Quais são as categorias com maior e menor receita dos últimos 12 meses?
categoria_maior_receita = receita_por_categoria.loc[receita_por_categoria['price'].idxmax()]
categoria_menor_receita = receita_por_categoria.loc[receita_por_categoria['price'].idxmin()]

print("Categoria com maior receita nos últimos 12 meses:")
print(categoria_maior_receita['product_category_name'], '-', categoria_maior_receita['price'])

print("\nCategoria com menor receita nos últimos 12 meses:")
print(categoria_menor_receita['product_category_name'], '-', categoria_menor_receita['price'])
```

```python
# Criar uma figura e um eixo (subplot)
fig, ax = plt.subplots(figsize=(7, 5))

# Plotar as categorias com maior e menor receita
categorias = ['Maior receita', 'Menor receita']
nomes_categorias = [categoria_maior_receita['product_category_name'], categoria_menor_receita['product_category_name']]
valores = [categoria_maior_receita['price'], categoria_menor_receita['price']]
cores = ['blue', 'red']

ax.bar(nomes_categorias, valores, color=cores)

# Adicionar os valores acima das barras
for i in range(len(categorias)):
    ax.text(nomes_categorias[i], valores[i], f'R$ {valores[i]:.2f}', ha='center', va='bottom')

# Adicionar título e rótulos aos eixos
ax.set_title('Receita das Categorias nos Últimos 12 Meses (Maior x Menor)')
ax.set_ylabel('Receita (em R$)')
ax.set_xlabel('Categoria')

# Mostrar o gráfico
plt.show()
```

## Top 10 maiores sellers (com maior receita)

```python
# Mesclar os conjuntos de dados itens_pedido e vendedores pelo seller_id
merge_vendedor = pd.merge(itens_pedido, vendedores, on='seller_id')

# Calcular a receita total de cada vendedor
receita_por_vendedor = merge_vendedor.groupby('seller_id')['price'].sum().reset_index()

# Remover valores nulos ou em branco da coluna de receita
receita_por_vendedor = receita_por_vendedor.dropna(subset=['price'])

# Classificar os vendedores com base na receita em ordem decrescente
top_sellers = receita_por_vendedor.sort_values(by='price', ascending=False).head(10)

top_sellers.head(10)
```

```python
# Ordenar os dados em ordem decrescente
top_sellers = top_sellers.sort_values('price', ascending=True)

# Visualizar os top 10 sellers com maior receita
plt.figure(figsize=(12,5))
bars = plt.barh(top_sellers['seller_id'], top_sellers['price'], color='skyblue')

# Adicionar os valores de receita em cada barra
for bar, value in zip(bars, top_sellers['price']):
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f'{value:.2f}',
             va='center', ha='right', color='black', fontsize=8, fontweight='bold')

plt.title('Top 10 Maiores Vendedores')
plt.xlabel('Vendedor')
plt.ylabel('Receita Total')
plt.xticks(rotation=75)

plt.show()
```

## Top 10 piores sellers (com menor receita)

```python
# Mesclar os conjuntos de dados itens_pedido e vendedores pelo seller_id
merge_vendedor = pd.merge(itens_pedido, vendedores, on='seller_id')

# Calcular a receita total de cada vendedor
receita_por_vendedor = merge_vendedor.groupby('seller_id')['price'].sum().reset_index()

# Remover valores nulos ou em branco da coluna de receita
receita_por_vendedor = receita_por_vendedor.dropna(subset=['price'])

# Classificar os vendedores com base na receita em ordem crescente
top_sellers_false = receita_por_vendedor.sort_values(by='price', ascending=True).head(10)

top_sellers_false.head(10)
```

```python
# Ordenar os dados em ordem crescente
top_sellers_false = top_sellers_false.sort_values('price', ascending=False)

# Visualizar os top 10 sellers com menor receita
plt.figure(figsize=(12,5))
bars = plt.barh(top_sellers_false['seller_id'], top_sellers_false['price'], color='orange')

# Adicionar os valores de receita em cada barra
for bar, value in zip(bars, top_sellers_false['price']):
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f'{value:.2f}',
             va='center', ha='right', color='black', fontsize=8, fontweight='bold')

plt.title('Top 10 Piores Vendedores')
plt.xlabel('Receita Total')
plt.ylabel('Vendedor')

plt.show()
```

## Existem sellers que vendem o mesmo produto? se sim, quais são? qual é a variação de preço praticado entre os sellers?

```python
# Calcular a quantidade de vendedores para cada produto e decrescente
qtd_vendedores = (merge_vendedor.groupby('product_id')['seller_id'].nunique().reset_index()).sort_values(by='seller_id', ascending=False)
qtd_vendedores.columns = ['product_id', 'qtd_sellers']

# Merge com o dataset original para obter os preços dos produtos
qtd_vendedores_precos = qtd_vendedores.merge(merge_vendedor, on='product_id', how='inner')

# Agrupar por produto e calcular o preço mínimo e máximo praticado pelos vendedores, bem como a diferença entre eles
resumo_precos = qtd_vendedores_precos.groupby('product_id')['price'].agg([('preco_minimo', 'min'), ('preco_maximo', 'max')])
resumo_precos['diferenca_preco'] = resumo_precos['preco_maximo'] - resumo_precos['preco_minimo']

# Merge com o resumo dos preços
resumo_final = qtd_vendedores.merge(resumo_precos, on='product_id', how='inner')

# Exibir os produtos com mais vendedores
print("Resumo dos Preços dos Produtos com Mais Vendedores:")
resumo_final.head(10)
```

```python
# Selecionar os 10 primeiros produtos com base na quantidade de vendedores e ordenar em ordem decrescente
resumo_final_classif = resumo_final.nlargest(10, 'qtd_sellers').sort_values(by='qtd_sellers', ascending=True)

# Criar um gráfico de barras horizontais com barras paralelas para preço mínimo e máximo e quantidade de vendedores em cada produto
plt.figure(figsize=(12, 8))
plt.barh(resumo_final_classif['product_id'], resumo_final_classif['preco_maximo'], color='orange', label='Preço Máximo')
plt.barh(resumo_final_classif['product_id'], resumo_final_classif['preco_minimo'], color='skyblue', label='Preço Mínimo')
plt.barh(resumo_final_classif['product_id'], resumo_final_classif['qtd_sellers'], color='green', label='Qtd de Vendedores')

# Adicionar os valores dos preços mínimo e máximo em cada barra
for i, (min_price, max_price) in enumerate(zip(resumo_final_classif['preco_minimo'], resumo_final_classif['preco_maximo'])):
    plt.text(min_price, i, f'{min_price:.2f}', ha='right', va='center', color='black', fontsize=8)
    plt.text(max_price, i, f'{max_price:.2f}', ha='left', va='center', color='black', fontsize=8)

# Adicionar os valores da quantidade de vendedores em cada barra
for i, qtd in enumerate(resumo_final_classif['qtd_sellers']):
    plt.text(qtd, i, f'{qtd}', ha='left', va='center', color='black', fontsize=8)

# Configurar o título e rótulos dos eixos
plt.xlabel('Qtd de Vendedores / Preço Mínimo e Máximo')
plt.ylabel('Produto')
plt.title('Preço Mínimo e Máximo para Produtos com Mais Vendedores')
plt.legend()

# Mostrar o gráfico
plt.show()
```

## Ouve inflação no preço dos produtos ao passar do tempo? Se ouve, de quanto foi a variação desta inflação em % e em R$?

```python
# Converter a coluna 'order_purchase_timestamp' para o formato datetime
pedidos['order_purchase_timestamp'] = pd.to_datetime(pedidos['order_purchase_timestamp'])

# Extrair o ano e mês da coluna 'order_purchase_timestamp'
pedidos['year_month'] = pedidos['order_purchase_timestamp'].dt.to_period('M')

# Mesclar os dados de pedidos e itens de pedido
merge_pedidos_itens = pd.merge(pedidos, itens_pedido, on='order_id', how='inner')

# Calcular a média dos preços para cada mês
precos_medios_mensais = merge_pedidos_itens.groupby('year_month')['price'].mean()

# Calcular a variação percentual dos preços entre meses consecutivos
variacao_preco = precos_medios_mensais.pct_change() * 100

# Calcular estatísticas personalizadas da variação percentual
variacao_media = variacao_preco.mean()
desvio_padrao = variacao_preco.std()
variacao_max = variacao_preco.max()
variacao_min = variacao_preco.min()

# Exibir estatísticas descritivas e personalizadas da variação percentual
print('\n...................................................')
print('Variação Mensal')
print('...................................................')
print(variacao_preco)

print('\n...................................................')
print("Estatísticas Personalizadas da Variação Percentual:")
print('...................................................')
print(f"Média: {variacao_media:.2f}%")
print(f"Desvio Padrão: {desvio_padrao:.2f}%")
print(f"Máximo: {variacao_max:.2f}%")
print(f"Mínimo: {variacao_min:.2f}%")
```

```python
# Converter os índices do tipo 'Period' para 'string'
variacao_preco.index = variacao_preco.index.astype(str)

# Criar um gráfico para visualizar a variação percentual dos preços ao longo do tempo
plt.figure(figsize=(10, 6))
plt.plot(variacao_preco.index, variacao_preco.values, marker='o', linestyle='-')
plt.xlabel('Ano-Mês')
plt.ylabel('Variação Percentual (%)')
plt.title('Variação Percentual dos Preços dos Produtos ao Longo do Tempo')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Top 10 melhores sellers (com mais reviews positivas) e Top 10 piores sellers (com mais reviews negativas)?

```python
# Mesclar os conjuntos de dados itens_pedido e comentarios_pedido pelo order_id
merge_review = pd.merge(itens_pedido, comentarios_pedido, on='order_id')

# Calcular a média dos scores de cada seller
media_scores_sellers = merge_review.groupby('seller_id')['review_score'].mean().reset_index()

# Definir critérios para reviews boas (score >= 4) e ruins (score <= 2)
boas = media_scores_sellers[media_scores_sellers['review_score'] >= 4]
ruins = media_scores_sellers[media_scores_sellers['review_score'] <= 2]

# Top 10 melhores sellers
top_melhores_sellers = boas.sort_values(by='review_score', ascending=False).head(10)

# Top 10 piores sellers
top_piores_sellers = ruins.sort_values(by='review_score', ascending=True).head(10)

print("Top 10 Melhores Sellers: ===========================")
print(top_melhores_sellers)
print("\nTop 10 Piores Sellers: =============================")
print(top_piores_sellers)
```

```python
# Criar gráfico dos top 10 melhores sellers
plt.figure(figsize=(12, 6))
plt.barh(top_melhores_sellers['seller_id'], top_melhores_sellers['review_score'], color='green')
plt.xlabel('Média de Scores')
plt.ylabel('ID do Seller')
plt.title('Top 10 Melhores Sellers (com mais reviews positivas)')
plt.gca().invert_yaxis()  # Inverter a ordem dos sellers
plt.show()

# Criar gráfico dos top 10 piores sellers
plt.figure(figsize=(12, 6))
plt.barh(top_piores_sellers['seller_id'], top_piores_sellers['review_score'], color='red')
plt.xlabel('Média de Scores')
plt.ylabel('ID do Seller')
plt.title('Top 10 Piores Sellers (com mais reviews negativas)')
plt.gca().invert_yaxis()  # Inverter a ordem dos sellers
plt.show()
```

## Existe relação entre a quantidade de vendas e a quantidade de reviews para os sellers? É possível identificar aumento ou queda na venda de um seller com base nas avaliações do que já foi vendido no passado?



```python
# Calcular a quantidade de vendas para cada seller
qtd_vendas_seller = merge_pedidos_itens['seller_id'].value_counts().reset_index()
qtd_vendas_seller.columns = ['seller_id', 'quantidade_vendas']

# Calcular a quantidade de reviews para cada seller
qtd_reviews_seller = merge_review['seller_id'].value_counts().reset_index()
qtd_reviews_seller.columns = ['seller_id', 'quantidade_reviews']

# Combina os dados de quantidade de vendas e quantidade de reviews por seller
qtd_vendas_reviews = qtd_vendas_seller.merge(qtd_reviews_seller, on='seller_id', how='inner')

# Calcular o coeficiente de correlação entre quantidade de vendas e quantidade de reviews
correlacao = qtd_vendas_reviews['quantidade_vendas'].corr(qtd_vendas_reviews['quantidade_reviews'])

print("Coeficiente de correlação entre quantidade de vendas e quantidade de reviews:", correlacao)
```

```python
# Criar o gráfico de dispersão
plt.figure(figsize=(10, 6))
plt.scatter(qtd_vendas_reviews['quantidade_vendas'], qtd_vendas_reviews['quantidade_reviews'], color='blue', alpha=0.5)
plt.xlabel('Quantidade de Vendas')
plt.ylabel('Quantidade de Reviews')
plt.title('Relação entre Quantidade de Vendas e Quantidade de Reviews por Seller')
plt.grid(True)
plt.show()
```

## *Tema livre:* A sazonalidade das vendas ao longo do ano.

```python
# Converter a coluna 'order_purchase_timestamp' para o formato datetime
merge_pedidos_itens['order_purchase_timestamp'] = pd.to_datetime(merge_pedidos_itens['order_purchase_timestamp'])

# Extrair o ano e mês da coluna 'order_purchase_timestamp'
merge_pedidos_itens['ano_mes'] = merge_pedidos_itens['order_purchase_timestamp'].dt.to_period('M')

# Agrupar as vendas por mês e calcular a soma das vendas para cada mês
vendas_por_mes = merge_pedidos_itens.groupby('ano_mes')['price'].sum()

print(vendas_por_mes)
```

```python
# Criar um gráfico de linha para visualizar as vendas ao longo do ano
plt.figure(figsize=(12, 6))
plt.plot(vendas_por_mes.index.astype(str), vendas_por_mes.values, marker='o', linestyle='-')
plt.xlabel('Ano-Mês')
plt.ylabel('Total de Vendas (R$)')
plt.title('Total de Vendas ao Longo do Ano')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
